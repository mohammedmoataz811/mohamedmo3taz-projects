```python
data.columns = (
    data.columns
    .str.strip()                
    .str.replace('\n', ' ')     
    .str.replace('"', '')       
    .str.replace(' ', '_')      
)

print(data.columns.tolist())
```

```python
def recommend_food(max_calories=500, min_protein=20):
    filtered = data[(data['Calories'] <= max_calories) & (data['Protein_(g)'] >= min_protein)]
    return filtered[['Company', 'Item', 'Calories', 'Protein_(g)', 'Total_Fat_(g)']]

recommend_food(500, 20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grouped = data.groupby('Company', as_index=False)[['Calories','Total_Fat_(g)','Protein_(g)']].mean()
grouped.head()
```

```python
grouped['Score'] = grouped['Protein_(g)'] / (grouped['Calories'] + grouped['Total_Fat_(g)'])   #  Creating diet score and filtering meals

```

```python
grouped = grouped.sort_values('Score', ascending=True)
print(grouped[['Company','Score']].head(10))
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=grouped, x='Company', y='Score', palette='viridis')
plt.title('Best Restaurants for Diet (Higher Score = Better)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Restaurant')
plt.ylabel('Diet Score')
plt.tight_layout()
plt.show()
```
